```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

## chromedriver.exe && 定義 driver
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)


for idx in range(3384, len(pitcher_to_scrape)):
    ## get to the website
    driver.get("https://baseballsavant.mlb.com/")
    name = pitcher_to_scrape[idx]

    try:
        ## locate by ID
        search = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'player-auto-complete')))
        search.send_keys(name) ##change name
        time.sleep(2)

        if name == 'García, Luis' or name == 'Garcia, Luis':
            search.send_keys(Keys.ARROW_DOWN)
            time.sleep(1)
            if (idx == 1507) or (idx == 1077):
                search.send_keys(Keys.ARROW_DOWN)
                time.sleep(1)
                
        if name == 'Allen, Logan':
            search.send_keys(Keys.ARROW_DOWN)
            time.sleep(1)

        search.send_keys(Keys.RETURN)

        clickable = driver.find_element(By.XPATH, '//*[@id="statcast_nav_type_pitching"]')
        clickable.click()
        
        webtable_df = driver.find_elements(By.XPATH, '//*[@id="statcast_stats_pitching"]/table/tbody/tr/td')

    except:
        driver.quit()


    rows = int(len(webtable_df)/21)
    season_li_to_append = []

    for i in range(0, rows):    
        start = i*21
        season_to_append = False
        career_to_append = False

        for j in range(start, start + 21):
            
            if webtable_df[j].text == '2022':
                season_to_append = True

            if season_to_append == True:
                season_li_to_append.append(webtable_df[j].text)

    season_li_to_append.insert(0, name) ## change name

    pitcher_season_df.loc[len(pitcher_season_df)] = season_li_to_append
```
