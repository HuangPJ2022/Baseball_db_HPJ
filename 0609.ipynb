```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

## chromedriver.exe && 定義 driver
PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)

for idx in range(526, len(hitter_li)):
    ## get to the website
    driver.get("https://baseballsavant.mlb.com/")
    name = hitter_li[idx]

    try:
        ## locate by ID
        search = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'player-auto-complete')))
        search.send_keys(name) ##change name
        time.sleep(3)

        ## enter
        
        if name == 'Will Smith':
            search.send_keys(Keys.ARROW_DOWN)
            time.sleep(1)

        search.send_keys(Keys.RETURN)

        webtable_df = driver.find_elements(By.XPATH, '//*[@id="statcast_glance_batter"]/table/tbody/tr/td')

    except:
        driver.quit()


    rows = int(len(webtable_df)/19)
    season_li_to_append = []
    career_li_to_append = []

    for i in range(0, rows):    
        start = i*19
        season_to_append = False
        career_to_append = False

        for j in range(start, start + 19):
            
            if webtable_df[j].text == '2022':
                season_to_append = True
            
            if webtable_df[j].text == 'Player':
                career_to_append = True

            if season_to_append == True:
                season_li_to_append.append(webtable_df[j].text)

            if career_to_append == True:
                career_li_to_append.append(webtable_df[j].text)

    season_li_to_append.insert(0, name) ## change name

    hitter_season_df.loc[len(hitter_season_df)] = season_li_to_append

    ## 輸入
    """search.send_keys("Shohei")
    time.sleep(1)

    clickable = driver.find_element(By.XPATH, '//*[@id="player_search_container"]/div/div/div/div/button')
    clickable.click()"""
```
