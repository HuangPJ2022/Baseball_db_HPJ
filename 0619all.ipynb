```python
one_hot_encoded_data = pd.get_dummies(result_all, columns = ['pitch_type'])
one_hot_encoded_data

## 自動 drop pitch_type 生成 one hot encoder
```

```python
result_all = one_hot_encoded_data
result_all
```

```python
result_all = result_all.drop(columns=['umpire','player_name', 'des', 'Season', 'batter', 'pitcher'], axis=1)
```

```python
result_all = result_all.drop(columns=['fielder_2'], axis=1)
```

```python
result_all = result_all.drop(columns=['game_pk'], axis=1)
```

```python
result_all = result_all.drop(columns=['pitch_name'], axis=1)
```

```python
result_all
```

```python
score_differential = []

for idx in range(0, len(result_all)):
    home = result_all.iloc[idx]['home_score']
    away = result_all.iloc[idx]['away_score']
    differential = abs(home - away)

    score_differential.append(differential)

len(score_differential)
```

```python
for idx, col in enumerate(result_all.columns):
    print(idx, col)
```

```python
result_all.insert(18, 'score_differential', score_differential)
```

```python
result_all
```

```python
same_handed = []

for idx in range(0, len(result_all)):
    hitter_hand = result_all.iloc[idx]['stand']
    pitcher_hand = result_all.iloc[idx]['p_throws']

    if hitter_hand == pitcher_hand:
        same_handed.append(True)
    
    else:
        same_handed.append(False)

same_handed
```

```python
result_all.insert(8, 'same_handed', same_handed)
```
