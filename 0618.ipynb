```python
new_all_hitter_merge = pd.DataFrame(columns=['pitch_type', 'release_speed', 'release_pos_x', 'release_pos_z',
       'player_name', 'batter', 'pitcher', 'description', 'zone', 'des',
       'stand', 'p_throws', 'outs_when_up', 'inning', 'inning_topbot',
       'fielder_2', 'umpire', 'effective_speed', 'release_spin_rate',
       'release_extension', 'game_pk', 'release_pos_y', 'at_bat_number',
       'pitch_name', 'home_score', 'away_score', 'spin_axis',
       'delta_home_win_exp', 'delta_run_exp', 'Name', 'Season', 'Age', 'Pitches', 'Batted_Balls', 'Barrels',
       'Barrel%', 'Barrel/PA', 'Exit_Velocity', 'Max_EV', 'Launch_Angle',
       'Sweet_Spot%', 'XBA', 'XSLG', 'WOBA', 'XWOBA', 'XWOBACON', 'HardHit%',
       'K%', 'BB%'])

merge_counter = 0
for idx, who in enumerate(new_all['des']):
    temp_merge_li = []

    if who == new_hitter_season_df.iloc[merge_counter]['Name']:

        for info in new_all.iloc[idx]:
            temp_merge_li.append(info)

        for info2 in new_hitter_season_df.iloc[merge_counter]:
            temp_merge_li.append(info2)

        merge_counter += 1

        new_all_hitter_merge.loc[len(new_all_hitter_merge)] = temp_merge_li
    else:
        pass
```
